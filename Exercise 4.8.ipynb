# STEP 4: Creating a loyalty flag for existing customers

```python
#Determine the Maximum Order Number for Each Customer:
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform('max')
```

```python
df_ords_prods_merge.head()
```

```python
# Create the Loyalty Flag Based on the Maximum Order Number:
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge['loyalty_flag'].value_counts(dropna =False)
```

```python
 df_ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(100) 
```

# STEP 5: Determining whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
# Group by the loyalty_flag and calculate the basic statistics
loyalty_stats = df_ords_prods_merge.groupby('loyalty_flag')['prices'].agg(['mean', 'median', 'min', 'max']).reset_index()
print(loyalty_stats)
```

# Analysis of the results: 
Loyal Customers:Tend to spend less on average compared to regular and new customers. This could indicate that loyal customers might be more price-conscious or prefer products with lower prices.
Regular Customers:Spend moderately compared to new and loyal customers. They may not yet be fully committed to the brand but are more engaged than new customers.
New Customers:Have the highest average spending, which might be due to exploratory behavior, trying out new products without regard to cost

# STEP 6: This will create a column spending_flag in entire DataFrame, categorizing each user as either a "Low spender" or a "High spender" based on their average spending.


```python
# Calculate the mean prices per user
user_avg_prices = df_ords_prods_merge.groupby('user_id')['prices'].mean().reset_index()

# Create a new column for the spending flag
user_avg_prices['spending_flag'] = user_avg_prices['prices'].apply(lambda x: 'Low spender' if x < 10 else 'High spender')
```

```python
# Merge the spending flag back into the original dataframe
df_ords_prods_merge = df_ords_prods_merge.merge(user_avg_prices[['user_id', 'spending_flag']], on='user_id', how='left')

# Verify the first few rows
df_ords_prods_merge.head()
```

# STEP 7: Creating an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
# Calculate the median days since prior order per user
user_median_days = df_ords_prods_merge.groupby('user_id')['days_since_prior_order'].median().reset_index()

# Create a new column for the order frequency flag
user_median_days['order_frequency_flag'] = user_median_days['days_since_prior_order'].apply(
    lambda x: 'Non-frequent customer' if x > 20 else ('Regular customer' if 10 < x <= 20 else 'Frequent customer')
)

# Merge the order frequency flag back into the original dataframe
df_ords_prods_merge = df_ords_prods_merge.merge(user_median_days[['user_id', 'order_frequency_flag']], on='user_id', how='left')

# Verify the first few rows
df_ords_prods_merge.head()
```
